```python
# 2. Path to the dataset yaml
data_yaml = f"{dataset.location}/data.yaml"
print("Data YAML file is here:", data_yaml)
```

```python
from ultralytics import YOLO

model = YOLO("yolo11n.pt")

history=model.train(
    data=data_yaml,
    epochs=70,
    imgsz=640,
    batch=16,
    project="dentistry-crown-46s6n",
    name="yolov11_roboflow",
)
```

```python
# Validation
model.val(data=data_yaml)

# Prediction on validation images
model.predict(source=f"{dataset.location}/valid/images", conf=0.25)
```
